**1. Creating Dataset**

```python
import pandas as pd

data = {
    'student_id': range(1,16),
    'gender':['Male','Male', '','Male','Male','Female','Male','Male','Female','Male','Male','Female','Male','Male','Female'],
    'test_date': pd.date_range(start = '2024-02-14',periods=15),
    'study_hours':[5,2,5,2,6,3,1,5,3,4,2,4,6,4,3],
    'test_score':[80,75,87,98,75,83,87,90,93,90,83,87,90,93,90],
    'attendance':['Present','Absent','Absent','Absent','Present','Present','Present','Present','Present','Present','Present','Present','Absent','Absent','Present']
}

df = pd.DataFrame(data)

df.to_csv('student_performance.csv', index=False)
```

2. Data Loading and Initial Examination

```python
df = pd.read_csv('student_performance.csv')

print(df.head(6))

print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sne


summary_stats = df[['test_score','study_hours']].describe()

#Histogram
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sne.histplot(df['test_score'],bins=15,kde=True)
plt.title('Distribution of Test Score')

#Scatterplot
plt.subplot(1,2,2)
sne.scatterplot(data=df,x='study_hours', y='test_score')
plt.title('Scatter Plot of Study Hours vs Test Score')

#Scatterplot
plt.figure(figsize=(8,6))
sne.scatterplot(data,x='study_hours', y='test_score')
plt.title('Scatter Plot of Study Hours vs Test Score')
```

4. Data Visualization

```python
#Gender Distribution
plt.figure(figsize=(6,4))
sne.countplot(data=df,x='gender')
plt.title('Gender Distribution')

#Test Score Over Time
plt.figure(figsize=(10,6))
sne.lineplot(data=df,x='test_date',y='test_score')
plt.title('Trend in Test Score Over Time')

correlation_matrix =df [['study_hours','test_score', 'attendance']].corr()
sne.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
```
