```python
import pandas as pd #Data Manipulation
import numpy as np #Mathematical Operation
import matplotlib.pyplot as plt #Data Visualization
from sklearn.linear_model import LinearRegression #Data Modeling/Traning Data
```

```python
df = pd.read_csv('Lesson_4_student_grades_dataset 2.csv')
print(df)
```

```python
plt.scatter(df['Hours_of_Study'],df['Grades'])
plt.xlabel('Independent Variable (HoursOfStudy)')
plt.xlabel('Dependent Variable (Grades)')
plt.title('Scatter Plot of Data')
plt.show()
```

```python
x = df[['Hours_of_Study']]

y = df[['Grades']]
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
slope = model.coef_[0]
intercept = model.intercept_
print("Slope: ", slope)
print("Intercept: ", intercept)
```

```python
plt.scatter(x, y)
plt.plot(x,model.predict(x), color='purple')
plt.xlabel('Independent Variable (x)')
plt.xlabel('Dependent Variable (y)')
plt.title('Linear Regression')
plt.show()
```

```python
new_x = [[np.random.randint(1,6)]]
predicted_y = model.predict(new_x)
print("Value of Hours of Study:", new_x)
print("Predicted Grades:",predicted_y)
```
